DATA HANDLING

    three ways :

 if the null value is above 40% then delete the column.

 if the null value is less then 40% then handle the data .
 
 if the null value is very less(like 2 to 10) then delete the row and if the null value is 30 or 35 % and it the important row in data then we will not delete the rows or column  ,we will handle the data.

 

```python
avg_age=round(df['age'].mean()) #for numeric data we calculate mean 
print(avg_age)                  #for alphabatical data we calculate mode
```

```python
df['age'].fillna(avg_age,inplace=True) #fillna is used to replace to null value
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['embarked'],inplace=True) #dropna is used to drop the null values
```
