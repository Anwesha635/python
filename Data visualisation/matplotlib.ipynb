```python
import matplotlib.pyplot as plt
```

```python
#plotting the figure
fig, ax=plt.subplots(figsize=(10,6)) 

#setting the background color
fig.set_facecolor('lightpink')
ax.set_facecolor('lightblue')

#setting the title and labels
ax.set_title('simple line chart',fontsize=20,color='red')
ax.set_xlabel('X-Axis',fontsize=15,color='green')
ax.set_ylabel('Y-Axis',fontsize=15,color='green')

x=[1,2,3,4,5,6,7,8,9]
y=[15,24,56,45,36,26,67,45,10]

ax.plot(x,y,
        marker='o',
        markerfacecolor='red',
        markeredgecolor='black',
        markeredgewidth=2,
        markersize=8,
        color='green',
        linestyle=':')
ax.axhline(10,color='black',alpha=0.1)
ax.axvline(5,color='black',alpha=0.1)

ax.text(2.5,33,'sample text',fontsize=12,color='purple')
```
