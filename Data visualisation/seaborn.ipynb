```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
df
```

```python
fig=sns.histplot(df,x='age',color='red')
```

```python
fig=sns.countplot(df,x='pclass',hue='survived')
```

```python
sns.countplot(df,x='survived',hue='sex',palette='coolwarm')
plt.title('count of survived by gender')
plt.show()
```
